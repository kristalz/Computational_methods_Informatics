```python
# Replace all missing values 
data = pd.read_csv('C:/Users/kryst/OneDrive/Desktop/Fall 1st/BIS 634/BIS_634_HW/incd.csv',skiprows=8,na_values = ['***','N/A ',' N/A ',' N/A','*']) 
data
```

```python
# Remove the lines that are not inside the table
# From the table, we know the last state/territory is Puerto Rico
# First, let's find the index 

data[data['State'] == 'Puerto Rico(6)'].index
```

```python
# Then remove all the rows below that index row 
data.drop(data.tail(74-53).index,inplace=True) 
data
```

```python
print(" \nCount total NaN at each column in the DataFrame : \n\n",
      htn_data.isnull().sum())
```

```python
# Drop the fields where there are high percentages of missing values
htn_data = htn_data.drop(columns =['Hypertension Past Year', 'Hypercholemia Past Year', 'Diabetes Age', 'Diabetes Type', 'Smoking Frequency'])
htn_data.head()
```

```python
# Drop the rows with missing values and change all values to integers 
htn_data = htn_data.dropna().astype(int)
```

```python
htn_data.head()
```

```python
# Group the ages into discrete intervals 
htn_data['Age'] = pd.cut(x=htn_data['Age'], bins=[17,34,54,64,70], labels=['18-34','35-54', '55-64','64-70'])
htn_data['Age']
```

```python
columns_list = list(htn_data.columns)
columns_list
```

```python
htn_data = htn_data.astype('category') # The values in the data set are all categorical numbers
htn_data['Hypertension Diagnose'].replace({1:'Yes', 2:'No'}, inplace=True) # Replace the "Yes/No" values for 1 or 2. 
htn_data.head()
```

```python
# Compare counts of hypertension diagnose by ages, genders, and races
age_gender_race = htn_data.groupby(['Age','Gender','Race&Ethnicity'])['Hypertension Diagnose'].count()
age_gender_race = age_gender_race.reset_index(name='Hypertension')
age_gender_race['Gender'].replace({1:'Male', 2:'Female'}, inplace=True)
age_gender_race['Race&Ethnicity'].replace({1:'White, non-Hispanic', 2:'Black, non-Hispanic', 
                                    3:'Other, non-Hispanic', 4:'Hispanic'}, inplace=True)
(p9.ggplot(data=age_gender_race,
           mapping=p9.aes(x='Age',
                          y='Hypertension', fill = 'Gender'))
    + p9.geom_col(stat='identity', position='dodge', show_legend=True)  
    + p9.geom_text(p9.aes(label='Hypertension'),   
                position=position_dodge(width=0.9),                              
                size=8, va='bottom', format_string='{}')
    + p9.facet_wrap('Race&Ethnicity')
    + p9.ylim(0, 300)
)
```

Hypertension is more prevalent in Whites(non-Hispanic) and 35-54 age group. The numbers of hypertension diagnose are relatively the same among females and males. 

```python
# Compare counts of hyptersion diagnose by genders
Gender_counts = htn_data.groupby(['Hypertension Diagnose', 'Gender'])['Gender'].count()
Gender_counts = Gender_counts.reset_index(name='Counts')
Gender_counts['Gender'].replace({1:'Male', 2:'Female'}, inplace=True)
(p9.ggplot(data=Gender_counts,
           mapping=p9.aes(x='Hypertension Diagnose',
                          y='Counts', fill = 'Gender'))
    + p9.geom_col(stat='identity', position='dodge', show_legend=True)  
    + p9.geom_text(p9.aes(label='Counts'),   
                position=position_dodge(width=0.9),                              
                size=8, va='bottom', format_string='{}')
)
```

Hypertension is more prevalent in the male responders.

```python
# Compare counts of hyptersion diagnose by Age groups
Age_counts = htn_data.groupby(['Hypertension Diagnose', 'Age'])['Age'].count()
Age_counts = Age_counts.reset_index(name='Counts')
(p9.ggplot(data=Age_counts,
           mapping=p9.aes(x='Hypertension Diagnose',
                          y='Counts', fill = 'Age'))
    + p9.geom_col(stat='identity', position='dodge', show_legend=True) 
    + p9.coord_flip()
    + p9.geom_text(p9.aes(label='Counts'),   
                    position=position_dodge(width=0.9),                              
                    size=8, ha='left', format_string='{}')
)
```

```python
# Compare counts of hyptersion diagnose by education, employment, income, and region
ed_employ_income_region = htn_data.groupby(['Education','Employment','Income','Region'])['Hypertension Diagnose'].count()
ed_employ_income_region = ed_employ_income_region.reset_index(name='Hypertension')
ed_employ_income_region['Education'].replace({2:'HS graduate or less', 3:'Some college', 4:'BA or above'}, inplace=True)
ed_employ_income_region['Employment'].replace({1:'Paid employee', 2:'Self-employed', 
                                    3:'Temporary lay off', 4:'Looking for work', 5:'Retired', 6:'Disabled', 7:'Other'}, inplace=True)
ed_employ_income_region['Region'].replace({1:'Northeast', 2:'Midwest', 3:'South', 4:'West'},inplace=True)                                
(p9.ggplot(data=ed_employ_income_region,
           mapping=p9.aes(x='Education',
                          y='Hypertension', fill = 'Region'))
    + p9.geom_col(stat='identity', position='dodge', show_legend=True) 
    + p9.theme(axis_text_x=p9.element_text(angle=90))
    + p9.facet_wrap('Employment')
)
```

More employed repsonders are diagnosed with hypertension, especially in the South and Midwest regions. 

```python
# Facet by education levels 
(p9.ggplot(data=ed_employ_income_region,
           mapping=p9.aes(x='Employment',
                          y='Hypertension', fill = 'Region'))
    + p9.geom_col(stat='identity', position='dodge', show_legend=True) 
    + p9.theme(axis_text_x=p9.element_text(angle=90))
    + p9.facet_wrap('Education')
)
```

Again, more repsonders who are employed are diagnosed with hypertension in the South and Midwest regions. Hypertension seems to be more prevalent in those who have higher educational degrees. 

```python
# Compare counts of hypertension diagnose by prediabetes, diabetes, hypercholemia, smoking status, and levels of depression
other_conditions = htn_data.groupby(['Prediabetes', 'Diabetes', 'Hypercholemia Diagnose', 'Smoking', 'Depression'])['Hypertension Diagnose'].count()
other_conditions = other_conditions.reset_index(name='Hypertension')
other_conditions['Depression'].replace({1:'Daily', 2:'Weekly', 3:'Monthly', 4:'A few times a year', 5:'Never'}, inplace=True)
cols = ['Prediabetes', 'Diabetes', 'Hypercholemia Diagnose', 'Smoking']
for i in range(len(cols)):  
    other_conditions[cols[i]].replace({1:'Yes', 2:'No'}, inplace=True)
other_conditions.head()
```

```python
# Compare counts of hyptersion diagnose by activity levels, age, and employment status
activity = htn_data.groupby(['Age','Employment','Light/Moderate Activity Frequency','Muscular Activity Frequency','Vigorous Activity Frequency'])['Hypertension Diagnose'].count()
activity = activity.reset_index(name='Hypertension')
activity['Employment'].replace({1:'Paid employee', 2:'Self-employed', 
                                    3:'Temporary lay off', 4:'Looking for work', 5:'Retired', 6:'Disabled', 7:'Other'}, inplace=True)
cols = ['Light/Moderate Activity Frequency', 'Muscular Activity Frequency', 'Vigorous Activity Frequency']
for i in range(len(cols)):
    activity[cols[i]].replace({1:'Never', 2:'Daily', 3:'Weekly', 4:'Monly', 
                                5:'Yearly', 6:'Unable to do this activity'}, inplace=True)
activity.head()
```

```python
# Light/moderate activity
(p9.ggplot(data=activity,
           mapping=p9.aes(x='Light/Moderate Activity Frequency',
                          y='Hypertension', fill = 'Employment'))
    + p9.geom_col(stat='identity', position='dodge', show_legend=True) 
    + p9.theme(axis_text_x=p9.element_text(angle=90)) 
    + p9.facet_wrap('Age')
)
```

Most responders have at least 10 min light to moderate physical activities weekly. However, the relationship between activity frequency and hypertension is not clear. Hypertension is most prevalent in those who are employeed before age of 64. 

```python
# Muscular Activity
(p9.ggplot(data=activity,
           mapping=p9.aes(x='Muscular Activity Frequency',
                          y='Hypertension', fill = 'Employment'))
    + p9.geom_col(stat='identity', position='dodge', show_legend=True) 
    + p9.theme(axis_text_x=p9.element_text(angle=90)) 
    + p9.facet_wrap('Age')
)
```

```python
# Vigorous Activity
(p9.ggplot(data=activity,
           mapping=p9.aes(x='Vigorous Activity Frequency',
                          y='Hypertension', fill = 'Employment'))
    + p9.geom_col(stat='identity', position='dodge', show_legend=True) 
    + p9.theme(axis_text_x=p9.element_text(angle=90)) 
    + p9.facet_wrap('Age')
)
```

From the three graphs above, we can see that the relationship between activity frequency and hypertension is not clear, regardless of the type of activity.
